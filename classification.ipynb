```python
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_text as text
import pandas as pd
```

```python
df=pd.read_csv('spam.csv')
```

```python
df_spam=df[df['Category']=='spam']
df_ham = df[df['Category']=='ham']
df_downsample=df_ham.sample(df_spam.shape[0])
df_balace=pd.concat([df_spam,df_downsample])
df_balace['Category'].value_counts()
```

```python
df_balace['spam']=df_balace['Category'].apply(lambda x: 1 if x=='spam' else 0)
```

```python
X=df_balace['Message'].values
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,df_balace['spam'].values, stratify=df_balace['spam'].values)
```

```python
bert_process=hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3")
```

```python
bert_encoder = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4")
```

```python
def  get_embeddings(data):
    embeds=bert_process(data)
    #print(embeds)
    print(bert_encoder(embeds))
    return bert_encoder(embeds)['pooled_output'] 

get_embeddings(['hey everyone','hello all'])
```

```python
o=get_embeddings(['mango','hell','elon','banana'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity([o[1]],[o[2]])
```

```python
from tensorflow.keras.models import Model
```

```python
from tensorflow.keras.layers import Input,Dropout,Dense
```

```python
inputs=Input(shape=(),dtype=tf.string,name='text')
preprocessed_text=bert_process(inputs)
encoder=bert_encoder(preprocessed_text)


drop=Dropout(0.2,name='dropout')(encoder['pooled_output'])
output=Dense(2,activation='softmax',name='outut')(drop)


```

```python
model=Model(inputs=[inputs],outputs=[output])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss=tf.keras.losses.CategoricalCrossentropy(),metrics=[tf.keras.metrics.Accuracy(),tf.keras.metrics.Precision(),tf.keras.metrics.Recall()])
```

```python
y_train=tf.keras.utils.to_categorical(y_train)
```

```python
model.fit(X_train,y_train,epochs=10)
```

```python
model.evaluate(X_test, tf.keras.utils.to_categorical(y_test))
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
import numpy as np
y_pred=[]
for i in range(len(pred)):
    y_pred.append(np.argmax(pred[i]))
```

```python
y_pred=np.array(y_pred)
```

```python
y_pred
```

```python
print(classification_report(y_test,y_pred))
#y_pred.shape
```
